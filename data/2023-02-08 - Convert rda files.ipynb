```python
for desc in parts:
    #desc=parts[0]
    desc=desc.strip()
    desc=desc.replace("#' ","")
    desc=desc.replace("#'","")

    lines=desc.split('\n')
    newlines=[]
    for i,line in enumerate(lines):
        if i==0:
            newlines.append(f"# {line}")
            continue

        if "@format" in line:
            newlines.append(f"### {line[8:]}")
            continue

        if r"\describe" in line:
            newlines.append("")
            continue

        if line.startswith("}"):
            newlines.append("")
            continue

        if r"\item" in line:
            part1=line.split('{')[1].split('}')[0]
            part2=line.split('{')[2].split('}')[0]

            newlines.append(f"- **{part1}**: {part2}")
            continue

        if "@source" in line:
            newlines.append(f"### Source: {line[8:]}")
            continue

        if line==lines[-1]:  # last line is the source file
            newlines.append(f"### Filename: {line[1:-1]}.csv from https://github.com/adamtclark/gauseR")
            savefname=line[1:-1]+".md"
            continue


        newlines.append(line)


    desc='\n'.join(newlines)

    print(savefname)


    with open(savefname,"w") as fid:
        fid.write(desc)
```
